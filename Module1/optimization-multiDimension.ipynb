# Optimization - Multi-Dimensional

Multidimensional optimization involves finding the minimum or maximimum of a function of more than one variable. The minimization is done using the built-in function [`fmin`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html) from the Python SciPy package. The maximum is handled by multiplying the function by $-1$ and using the `fmin` as discussed in [`optimization-goldenSearch.ipynb`](https://colab.research.google.com/drive/1e9rQHNJGuxXb1ne9b8P0Yjd2nvwH07BQ?usp=sharing).

### Example:
Find the minumum value and location of $f(x,y)=2+x-y+2x^2+2xy+y^2$ in the intervals $-2 \le x \le 0$ and $0 \le y \le 3$.

```
from scipy.optimize import fmin
import numpy as np
import matplotlib.pyplot as plt

x0 = [-0.5, 0.5] # initial guess
tol = 1e-6 # error tolerance
f = lambda x: 2 + x[0] - x[1] + 2 * x[0]**2 + 2 * x[0] * x[1] + x[1]**2   # function definition f(x) where x is a vector: x[0]->x and x[1]->y
sol = fmin(f, x0, xtol=tol)
print('x coordinate of min location = ',sol[0])
print('y coordinate of min location = ',sol[1])
print('min value of f at xmin and ymin = ',f(sol))

# Check with the 3D plot of the function
x=np.linspace(-2,0)
y=np.linspace(0,3)
xv, yv = np.meshgrid(x, y)
zv=2+xv-yv+2*xv**2+2*xv*yv+yv**2  # z data based on x and y according to the function
fig = plt.figure()
ax = fig.gca(projection='3d')   
surf = ax.plot_surface(xv,yv,zv)  # 3D surfce plot        
plt.xlabel('x')
plt.ylabel('y')
ax.set_zlabel('z')   # z axis label
plt.xticks(np.arange(-2, 0, step=0.5))  #xtick marks adjust
plt.yticks(np.arange(0, 3, step=0.5))   #ytick marks adjust
ax.view_init(40,20) # display angle of the 3D plot
```
