# Basic statistics functions

* [mean](https://numpy.org/doc/stable/reference/generated/numpy.mean.html): the arithmetic mean (average) of the data.
* [median](https://numpy.org/doc/stable/reference/generated/numpy.median.html#numpy.median): middle number when the list is sorted.
* [sort](https://numpy.org/doc/stable/reference/generated/numpy.sort.html): sorts a list.
* [mode](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mode.html): returns the most common number in the list and the number of times it occurs. It only returns the first most-occured value (there might be other numbers with the same number of occurance).
* [min](https://numpy.org/doc/stable/reference/generated/numpy.amin.html#numpy.amin)/[max](https://numpy.org/doc/stable/reference/generated/numpy.amax.html#numpy.amax): returns min/max of an array.
* [sum](https://numpy.org/doc/stable/reference/generated/numpy.sum.html#numpy.sum): sum of data in the list
* Vairance and Standard deviation: in the functions in python, the degrees of freedom is `N-ddof`, where `ddof` is the input aurgument to the functions. By default, ddof=0. We need to set `ddof=1` when using these functions:
  * [var](https://numpy.org/doc/stable/reference/generated/numpy.var.html#numpy.var): variance of data.
  * [std](https://numpy.org/doc/stable/reference/generated/numpy.std.html#numpy.std): standard deviation of data.

```
import numpy as np
A = np.array([1.,2.,3.,4.,5.,5.,3.,7., 8., 9., 3., 4., 2., 7., 3.])
print('mean = ', np.mean(A))      # mean
print()
print('median= ', np.median(A))   # median
print(np.sort(A))   # the list sorted, it is seen that 4 is the median
print()
from scipy.stats import mode
print('mode=',mode(A))
print()
print('min = ', np.min(A))      # min
print('max = ', np.max(A))      # max
range = np.max(A) - np.min(A)   # range od data
print('range=', range)          
print()
print('sum = ', np.sum(A))      # sum of element
print()
# for the variance and standard deviation, we have use the paramter ddof=1 to set the degrees of freedom to N-1. 
# Otherwise, by default it is ddof=0 which give N degrees of freedom.
print('variance =', np.var(A, ddof=1))               # variance
print('standard deviation =', np.std(A, ddof=1))     # standard deviation
```

# Histogram

The histrogram provides a visual presentation of frequency of occurance of data within the list. It It also shows the extent of spread of the data.

* [hist](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.hist.html): plot a histogram and returns the locations of bins (which are the sub-intervals containing the data) and how many data points they contain.

```
import matplotlib.pyplot as plt
# histrogram with default number of bins
b = plt.hist(A)      
plt.xlabel('data')
plt.ylabel('frequency of occurance')
plt.title('histogram with default bins')
plt.show();
# printing the information about bins of the histogram
print('number of data points within each bin =', b[0])
print('edge of the bins =', b[1])
# We can specify the number of bins of the histogram
# we can also do other customizations such as changing the color
b = plt.hist(A, bins = 20 , color='r') 
plt.xlabel('data')
plt.ylabel('frequency of occurance')
plt.title('histogram with bins=20')
plt.show();
# printing the information about bins of the histogram
print('number of data points within each bin =', b[0])
print('edge of the bins =', b[1])
```

# Random numbers

* [randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.RandomState.random_integers.html#numpy.random.RandomState.random_integers): Returns a set of uniformly distributed random integer numbers N such that a <= N < b.
* [random](https://numpy.org/doc/stable/reference/random/generated/numpy.random.random.html#numpy.random.random): Returns a set of uniformly distributed random floating point numbers N such that 0.0 <= N < 1.0.
* [uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.RandomState.uniform.html#numpy.random.RandomState.uniform): Returns a random floating point numbers N such that a <= N < b.
* [standard_normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.RandomState.standard_normal.html#numpy.random.RandomState.standard_normal): generates standard Normal distribution (mean=0, stdev=1).

```
import numpy.random as rn
#generate 10 uniformly distributed random integer numbers between 21 (inclusive) and 35 (exclusive)
R = rn.randint(21, 35, 10)   
print('random numbers =',R)
```

```
import numpy.random as rn
#generate 10 uniformly distributed random numbers between 0 (inclusive) and 1 (exclusive)
R = rn.random(10)
print('random numbers =',R)
# Checking the random number distribution
plt.hist(R, color='g')   
plt.xlabel('random numbers')
plt.ylabel('frequency of occurance')
plt.title('uniformly distributed 10 random numbers')
plt.show();

# As the number of random numbers increase the plot becomes more and more uniform
R = rn.random(1000)   #generate 1000 uniformly distributed random numbers
# Checking the random number distribution
import matplotlib.pyplot as plt
plt.hist(R, color='m')   
plt.xlabel('random numbers')
plt.ylabel('frequency of occurance')
plt.title('uniformly distributed 1000 random numbers')
plt.show();
```

```
import numpy.random as rn
#generate 20 uniformly distributed random numbers between 80 (inclusive) and 100 (exclusive)
R = rn.uniform(80, 100,20)
print('random numbers =',R)
# Checking the random number distribution
import matplotlib.pyplot as plt
plt.hist(R, color='c')   
plt.xlabel('random numbers')
plt.ylabel('frequency of occurance')
plt.title('uniformly distributed 10 random numbers')
plt.show();
```

```
import numpy.random as rn
#generate 500 random numbers with normal distribution (mean=0, std=1)
R = rn.standard_normal(500)   
# print('random numbers =',R)
# Checking the random number distribution
import matplotlib.pyplot as plt
plt.hist(R, bins=20, color='b')   
plt.xlabel('random numbers')
plt.ylabel('frequency of occurance')
plt.title('500 random numbers with normal distribution')
plt.show();
```

```
import numpy.random as rn
#generate 500 random numbers with normal distribution (mean=20, std=5)
R = rn.normal(20,5,500)   
import matplotlib.pyplot as plt
plt.hist(R, bins=20, color='r')   
plt.xlabel('random numbers')
plt.ylabel('frequency of occurance')
plt.title('500 random numbers with normal distribution')
plt.show();
```
