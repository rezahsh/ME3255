# Polynomials

In many applications we are interested to fit a polynomial a $n^{th}$ order polynomial
$$y = p_n(x)=p_0x^n+p_1x^{n-1}+\ldots+p_{n-1}x+p_n$$
to our data, where $p$'s are the polynomial coefficients. These coefficients are to be determined by curve fitting using a set of $(x,y)$ data points.

# `polyfit` function
The [`polyfit`](https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html) function provides the polynomial coefficients by curve fitting. As an example, let's consider the following data points. Fitting a straight line to the data is as follows (note that `polyfit` returns the coefficient from $p_0$ to $p_n$ in order).

```
import numpy as np
x = np.array([10.,20.,30.,40., 50., 60., 70., 80.])
y = np.array([25., 70., 380., 550., 610., 1220., 830., 1450.])

p1=np.polyfit(x,y,1)       # 1st order polynomial (straight line)
print('Polynomial coefficients: ',p1) # printing the coefficients
```

Now, let's fit a second order polynomial:

```
p2=np.polyfit(x,y,2)       # 2nd order polynomial (straight line)
print('Polynomial coefficients: ',p2) # printing the coefficients
```

# `polyval` function
To evaluate the polynomial at a point $x_0$ we can use the function [`polyval`](https://numpy.org/doc/stable/reference/generated/numpy.polyval.html#numpy.polyval). If $x_0$ is an array, `polyval` returns an array of function values. This is particularly useful in plotting the polynomial.

```
x0 = 55 # a desired x value
y0 = np.polyval(p1, x0) # the polynomial value p4(x0)
print(' First order polynomial values at x=55:',y0)
# plotting 1st order polynomial
print('--------------------------------------')
print()
xp = np.linspace(np.min(x),np.max(x))  # high resolution x and y for plotting
yp = np.polyval(p1,xp)

import matplotlib.pyplot as plt
plt.plot(xp,yp,'-')
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('First order polynomial')
plt.show();
#
print()
yp = np.polyval(p2,xp)
plt.plot(xp,yp,'b-')
plt.plot(x,y,'ro')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Second order polynomial')
plt.show();
```
